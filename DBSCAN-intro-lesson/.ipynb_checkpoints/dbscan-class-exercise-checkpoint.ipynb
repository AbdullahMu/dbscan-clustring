# DBSCAN  (Class Exercise)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Load the data

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
# top 5 rows
iris.head()
```

### Identify target and features

```python
X = iris.iloc[:, :-1]
y = iris.species
```

### Standardize X

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xs = scaler.fit_transform(X)
```

### Explore the data to look for potential clusters - see any?

```python
# pair plot
sns.pairplot(X)
```

```python
# select 2 features plot to use it in clustering presentation
plt.scatter(X.petal_length, X.petal_width)
```

### Setup DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.7, min_samples=4)
dbscan.fit(Xs)
```

### Set labels

```python
labels = dbscan.labels_
labels
```

### Identify core samples

```python
# Extract core sample indices
print(dbscan.core_sample_indices_)
```

### Declare the number of clusters

```python
n_clusters_ = len(set(labels) - set([-1]))
print(n_clusters_)
```

### Count majority cluster labels according to class label 

```python
pd.Series(labels).value_counts()
```

### Visualize Clusters

```python
plt.scatter(Xs[:,2], Xs[:,3], c=labels , label= set(labels))
```

```python
clus = pd.DataFrame({'petal_length': iris['petal_length'],
              'petal_width': iris['petal_width'],
              'cluster' : labels})
```

```python
clus.head()
```

```python
sns.lmplot(data=clus, x='petal_length', y='petal_width', hue='cluster', fit_reg=False)
```

### Try more eps and min_samples

```python
eps_lst = np.linspace(start = 0.1, stop = 2, num=10)
min_pts_lst = range(2, 5)

for eps in eps_lst:
    for min_pts in min_pts_lst:
        dbscan = DBSCAN(eps=eps, min_samples=min_pts)
        dbscan.fit(X)
        clus = pd.DataFrame({'petal_length': iris['petal_length'],
              'petal_width': iris['petal_width'],
              'cluster' : dbscan.labels_})
        sns.lmplot(data=clus, x='petal_length', y='petal_width', hue='cluster', fit_reg=False)
        plt.title('DBScan clustering using eps={}, min_pts={}'.format(eps,min_pts))
        plt.show()
```

### Evaluate Clusters

```python
from sklearn.metrics import silhouette_score, completeness_score, homogeneity_score, v_measure_score
```

```python
# silhouette score
print(silhouette_score(Xs, labels))
```

```python
# homogeneity_score
print(homogeneity_score(y, labels))
```

```python
# completeness_score
print(completeness_score(y, labels))
```

```python
# v_measure_score
print(v_measure_score(y, labels))
```
